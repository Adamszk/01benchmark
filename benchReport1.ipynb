```python
import matplotlib.pyplot as plt
from datetime import datetime
import pandas as pd
```

```python
compilers = []
size = []
compiler2 = []
speed = []
```

```python
f = open('codesize.txt','r')
g = open('output_file.txt','r')
```

```python
for row in f:
    row = row.split(' ')
    compilers.append(row[0])
    size.append(int(row[1]))
for row in g:
    row = row.split(' ')
    compiler2.append(row[0])
    speed.append(float(row[1])) 
```

```python
plt.bar(compilers, size, color=['red','green','blue','yellow','cyan','brown','magenta','black','purple','pink']
        , label = 'Code Size')
plt.xticks(compilers, compilers, rotation='vertical')  
plt.xlabel('Compiler', fontsize = 12)
plt.ylabel('Size (bits)', fontsize = 12)
  
plt.title('Code Size', fontsize = 20)
plt.legend()
plt.show()
```

```python
with open('output_file.txt') as f:
    lines = f.readlines()
    x = [str(line.split()[0]) for line in lines]
    y = [float(line.split()[1]) for line in lines]
fig = plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(111) 
plt.xticks(rotation=90)

ax1.set_title("Processing Time")
ax1.set_xlabel('Code Language')
ax1.set_ylabel('Time (s)')
colors = ['red','green','blue','yellow','cyan','brown','magenta','black','purple','pink']
ax1.bar(x,y, width=0.7,color=colors)
axes = plt.gca()
axes.set_ylim([0,80])
fig1=plt.gcf()
plt.show()
plt.draw()
```
